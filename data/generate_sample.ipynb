```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# generate sample data
n_samples = 1000

X, y = make_blobs(n_samples=[n_samples, n_samples], random_state=0)

df = pd.DataFrame(np.append(X,np.transpose([y]),axis=1), columns=['a','b','class'])
display(df.head())
```

```python
# plot clusters
df.plot.scatter(x='a',y='b',c='class', colormap='viridis', alpha=.5)
```

```python
df.to_csv("sample_data.csv")
```
