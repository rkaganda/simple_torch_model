```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import pickle
```

```python
scalerfile = "scaler.sav"

df = pd.read_csv("train_out.csv", index_col=0)
```

```python
display(df.head())
display(df.tail())
```

```python
feature_columns = [c for c in df.columns if c.startswith('x_')]
label_columns = [c for c in df.columns if c.startswith('y_')]
```

```python
labels = pd.get_dummies(df[label_columns], drop_first=True)
```

```python
display(labels.head())
display(labels.tail())
```

```python
df[label_columns] = labels
display(df.head())
display(df.tail())
```

```python
display(df.describe())
```

```python
df_saved_columns = df.columns
scaler = MinMaxScaler()
print(scaler)
scaler.fit(df) 
df = scaler.transform(df)
pickle.dump(scaler, open(scalerfile, 'wb'))

df = pd.DataFrame(df)

display(df.head())
display(df.tail())
```

```python
df.columns = df_saved_columns

display(df.head())
display(df.tail())
```
